```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("Lesson_4_student_grades_dataset.csv")
data.head()
```

```python
plt.scatter(data['Hours_of_Study'],data['Attendance_Percentage'])
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.title('Student Grades')
plt.show()
```

```python
X = data[['Grades']] #independent variable
Y = data[['Hours_of_Study']] #dependent variable
```

```python
model = LinearRegression()

model.fit(X,Y)
```

```python
plt.scatter(X,Y)
plt.plot(X, model.predict(X), color = 'red')
plt.xlabel('Grades = (X)')
plt.ylabel('Hours_of_Study = (Y) ')
plt.show()
```

```python
slope = model.coef_[0]

intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)
```

```python
new_X = [[31.494329084007404]]

predicted_Y = model.predict(new_X)

print('X=',new_X)
print('Predicted Value of Y',predicted_Y)
```
